```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
df
```

```python
med = df.Age.median()
```

```python
med
```

```python
df=df.fillna(med)
```

```python
df
```

```python
input = df[['Pclass','Sex','Age','Fare']]
input
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
input['Sex_n'] = le.fit_transform(df.Sex)
input
```

```python
input = input.drop(['Sex'],axis='columns')
```

```python
input
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(input,df.Survived)
```

```python
test = pd.read_csv('test.csv')
test
```

```python
ti = test[['Pclass','Sex','Age','Fare']]
tle = LabelEncoder()
```

```python
ti['Sex_n'] = tle.fit_transform(ti.Sex)
ti
```

```python
ti = ti.drop('Sex',axis='columns')
ti
```

```python
medi = ti.Age.median()
import math
a = math.floor(medi)
ti.Age = ti.Age.fillna(a)
ti=ti.fillna(0)
ti
```

```python
p = model.predict(ti)
p
```

```python
ans = pd.read_csv('test.csv')
ans
```

```python
ans = ans.drop(['Pclass','Name','Sex','Age','SibSp','Parch','Ticket','Fare','Cabin','Embarked'],axis='columns')
ans
```
