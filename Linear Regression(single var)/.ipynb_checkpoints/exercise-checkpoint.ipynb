```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df_pd=pd.DataFrame(df)
df_pd.rename(columns = {'per capita income (US$)':'Income'}, inplace = True)
df.head()
```

```python
%matplotlib inline
plt.scatter(df.year,df.Income,color="Green",marker='x')
```

```python
rg = linear_model.LinearRegression()
rg.fit(df[['year']],df.Income)
```

```python
rg.predict([[2020]])
```

```python
plt.scatter(df.year,df.Income,color="Green",marker='x')
plt.plot(df.year,rg.predict(df[["year"]]),color="Orange")
```
